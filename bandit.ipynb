```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 1, 0.01)
a = [0.1, 0.2, 0.5, 1]
b = [0.1, 0.2, 0.5, 1]
for aa,bb in zip(a, b):
  plt.plot(x, beta.pdf(x, aa, bb), label='a = ' + str(aa) + ', b =  ' + str(bb))
  plt.legend()
plt.show()
```

```python
a = [1, 5, 20, 50, 1000]
b = [1, 5, 5, 100, 800]
for aa,bb in zip(a, b):
  print('a = ' + str(aa) + ', b =  ' + str(bb))
  temp_min = 1
  temp_max = 0
  for _ in range(10):
    temp = beta.rvs(aa, bb)
    print(temp)
    temp_min = min(temp_min, temp)
    temp_max = max(temp_max, temp)
  print("max_prob:" + str(temp_max) + "  min_prob: " + str(temp_min))
  print("\n")
```
